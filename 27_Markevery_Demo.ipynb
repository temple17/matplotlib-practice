```python
import numpy as np
import matplotlib.pyplot as plt

cases = [
    None,
    8,
    (30, 8),
    [16, 24, 32],
    [0, -1],
    slice(100, 200, 3), 
    0.1,
    0.4,
    (0.2, 0.4)

]

delta = 0.11
x = np.linspace(0, 10-2*delta, 200) + delta

y = np.sin(x) + 1.0 + delta
```

# Markevery with linear scales

```python
fig, axs = plt.subplots(3, 3, figsize=(10,6), constrained_layout=True)
for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f'markevery={markevery}')
    ax.plot(x, y, 'o', ls = '-', ms=4, markevery = markevery)
```

# Markevery with log scales

```python
fig, axs = plt.subplots(3, 3, figsize=(10,6), constrained_layout=True)
for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f'markevery={markevery}')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.plot(x, y, 'o', ls='-', ms=4, markevery=markevery)
```

# Markevery on zoomed plot

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 6), constrained_layout=True)
for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f'markevery={markevery}')
    ax.plot(x, y, 'o', ls='-', ms=4, markevery=markevery)
    ax.set_xlim((6, 6.7))
    ax.set_ylim((1.1, 1.7))
```

# Markevery on polar plots

```python
r = np.linspace(0, 3.0, 200)
theta = 2* np.pi * r

fig, axs = plt.subplots(3, 3, figsize=(10,6), constrained_layout=True, subplot_kw={'projection':'polar'})

for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f'markevery={markevery}')
    ax.plot(theta, r, 'o', ls='-', ms=4, markevery=markevery)
```
